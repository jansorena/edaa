```python
import pandas as pd
import matplotlib.pyplot as plt

# Replace 'your_data_file.csv' with the name of your CSV file


# Read the CSV file
data_avl = pd.read_csv('/home/jansorena/repos/edaa/Boletin_2/results/results_avl_insert.csv')
data_rb = pd.read_csv('/home/jansorena/repos/edaa/Boletin_2/results/results_rb_insert.csv')
data_sp = pd.read_csv('/home/jansorena/repos/edaa/Boletin_2/results/results_sp_insert.csv')
data_sl = pd.read_csv('/home/jansorena/repos/edaa/Boletin_2/results/results_sl_insert.csv')

# Extract the relevant columns
n_values_avl = data_avl['n']
mean_time_avl = data_avl['t_mean']

n_values_rb = data_rb['n']
mean_time_rb = data_rb['t_mean']

n_values_sp = data_sp['n']
mean_time_sp = data_sp['t_mean']

n_values_sl = data_sl['n']
mean_time_sl = data_sl['t_mean']

# Plot the mean times
plt.figure(figsize=(10, 6))
plt.plot(n_values_rb, mean_time_rb, label='Red-Black Tree', marker='o')
plt.plot(n_values_avl, mean_time_avl, label='AVL Tree', marker='o')
plt.plot(n_values_sp, mean_time_sp, label='Splay Tree', marker='o')
plt.plot(n_values_sl, mean_time_sl, label='Skip List', marker='o')

# Add labels and title
plt.xlabel('Number of Insertions (n)')
plt.ylabel('Mean Insertion Time (nanoseconds)')
plt.title('Mean Insertion Time vs. Number of Insertions')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Replace 'your_data_file.csv' with the name of your CSV file


# Read the CSV file
data_avl = pd.read_csv('/home/jansorena/repos/edaa/Boletin_2/results/results_avl_search.csv')
data_rb = pd.read_csv('/home/jansorena/repos/edaa/Boletin_2/results/results_rb_search.csv')
data_sl = pd.read_csv('/home/jansorena/repos/edaa/Boletin_2/results/results_sl_search.csv')
data_st_most = pd.read_csv('/home/jansorena/repos/edaa/Boletin_2/results/results_st_search_most.csv')
data_st_least = pd.read_csv('/home/jansorena/repos/edaa/Boletin_2/results/results_st_search_least.csv')
data_st_same = pd.read_csv('/home/jansorena/repos/edaa/Boletin_2/results/results_st_search_same.csv')

# Extract the relevant columns
n_values_avl = data_avl['n']
mean_time_avl = data_avl['t_mean']

n_values_rb = data_rb['n']
mean_time_rb = data_rb['t_mean']

n_values_sl = data_sl['n']
mean_time_sl = data_sl['t_mean']

n_values_sp_most = data_st_most['n']
mean_time_sp_most = data_st_most['t_mean']

n_values_sp_least = data_st_least['n']
mean_time_sp_least = data_st_least['t_mean']

n_values_sp_same = data_st_same['n']
mean_time_sp_same = data_st_same['t_mean']

# Plot the mean times
plt.figure(figsize=(10, 6))
plt.plot(n_values_rb, mean_time_rb, label='Red-Black Tree', marker='o')
plt.plot(n_values_avl, mean_time_avl, label='AVL Tree', marker='o')
plt.plot(n_values_sl, mean_time_sl, label='Skip List', marker='o')
plt.plot(n_values_sp_most, mean_time_sp_most, label='Splay Tree (Most Frequent)', marker='o')
plt.plot(n_values_sp_least, mean_time_sp_least, label='Splay Tree (Least Frequent)', marker='o')
plt.plot(n_values_sp_same, mean_time_sp_same, label='Splay Tree (Same Frequency)', marker='o')

# Add labels and title
plt.xlabel('Number of Searchs (n)')
plt.ylabel('Mean Insertion Time (nanoseconds)')
plt.title('Mean Search Time vs. Number of Insertions')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
